#### **Mengganti Tanda `'?'` dengan `NaN`**

---



Beberapa dataset menandai nilai kosong atau tidak diketahui dengan simbol `'?'`, bukan `NaN`. Karena Python mengenali NaN sebagai `missing value`, kita perlu mengganti tanda `'?'` menjadi `np.nan`. **Tujuannya**, agar Python dan library analisis bisa mengenali nilai kosong secara standar.

```python
# Mengganti tanda '?' dengan NaN
df.replace('?', np.nan, inplace=True)
```

#### **Konversi Kolom ke Tipe Numerik**

Terkadang kolom yang seharusnya bertipe angka (numerik) malah dibaca sebagai object (`string`) karena mengandung karakter non-numerik seperti `'?'`. Fungsi `pd.to_numeric`(`..., errors='coerce'`) mengubah kolom menjadi numerik, dan jika ada data yang tidak bisa dikonversi, akan otomatis dijadikan `NaN`.**Tujuannya** untuk membersihkan data numerik dari nilai tidak valid.

```python
# Konversi kolom ke tipe numerik jika perlu (gunakan coerce agar string jadi NaN)
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
df['Height'] = pd.to_numeric(df['Height'], errors='coerce')
```

Dalam proses **Exploratory Data Analysis (EDA)**, visualisasi data sangat penting untuk memahami pola, tren, dan anomali dalam dataset secara intuitif.

#### **Visualisasi Menggunakan Boxplot**

---



**Boxplot** adalah visualisasi statistik yang digunakan untuk melihat sebaran data dan mengidentifikasi outlier atau nilai ekstrem. Dalam boxplot, garis tengah kotak menunjukkan median, sementara sisi bawah dan atas kotak masing-masing menunjukkan kuartil 1 (`Q1`) dan kuartil 3 (`Q3`). Whiskers atau garis yang menjulur dari kotak menunjukkan rentang data yang masih dianggap wajar, sedangkan titik-titik di luar *whiskers* merepresentasikan nilai-nilai yang jauh dari distribusi utama. **Visualisasi** ini sangat berguna untuk mengenali apakah terdapat individu dengan nilai usia, tinggi badan, atau berat badan yang tidak lazim atau menyimpang secara signifikan dari kelompok data lainnya.

```python
# Boxplot
features = ['Age', 'Height', 'Weight']
colors = ['skyblue', 'lightgreen', 'salmon']
titles = ['Boxplot untuk Usia', 'Boxplot untuk Tinggi Badan', 'Boxplot untuk Berat Badan']

plt.figure(figsize=(12, 4))
for i, (col, color, title) in enumerate(zip(features, colors, titles), 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x=df[col], color=color)
    plt.title(title)

plt.tight_layout()
plt.show()
```

**Boxplot untuk Usia (Age)**
* Distribusi utama berada di rentang usia sekitar `17` hingga `30` tahun.

* Titik-titik di sebelah kanan box merupakan outlier (nilai ekstrem), yaitu individu dengan usia lebih dari 40 tahun hingga lebih dari 100 tahun.

* Outlier ini bisa disebabkan oleh:

 * Data input yang tidak realistis (misalnya usia `100+`),

 * Atau memang mewakili individu yang tidak umum di populasi (tapi perlu dicek
validitasnya).

**Boxplot untuk Tinggi Badan (Height)**
* Rentang tinggi badan sebagian besar berada antara `1.5` meter hingga `1.9` meter.

* Ada outlier ekstrem di atas `2.5` meter, yang secara biologis tidak wajar dan kemungkinan besar merupakan data salah (error input).

* Outlier seperti ini perlu dibersihkan atau dipertimbangkan untuk dihapus saat preprocessing, karena bisa mengganggu analisis atau model.

**Boxplot untuk Berat Badan (Weight)**
* Berat badan mayoritas berkisar antara sekitar `50` hingga `150` kg.

* Terdapat outlier di atas 350 kg, yang sangat jarang dalam populasi normal dan kemungkinan besar merupakan data anomali.

* Sama seperti tinggi badan, outlier berat badan ekstrem harus dianalisis lebih lanjut (bisa jadi data salah atau kasus khusus).


#### **Visualisasi Distribusi Data**

---



**Histplot** digunakan untuk melihat distribusi frekuensi dari data numerik dengan menampilkan histogram, yaitu jumlah frekuensi pada setiap rentang nilai. Dengan menambahkan parameter `kde=True`, visualisasi ini juga menyertakan kurva estimasi kepadatan (kernel density estimation) yang memberikan gambaran lebih halus tentang bentuk distribusi data. **Tujuan utama dari histplot** adalah untuk mengetahui karakteristik sebaran data, apakah normal, skewed (miring ke kiri atau kanan), atau bahkan multimodal (memiliki lebih dari satu puncak). Visualisasi ini membantu menjawab pertanyaan seperti apakah berat badan peserta terkonsentrasi di nilai tertentu atau apakah usia tersebar merata dalam populasi.

```python
plt.figure(figsize=(15, 5))

# Distribusi Usia
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title('Distribusi Usia')
plt.xlabel('Usia')
plt.ylabel('Jumlah')

# Distribusi Berat Badan
plt.subplot(1, 3, 2)
sns.histplot(df['Weight'], bins=30, kde=True, color='green')
plt.title('Distribusi Berat Badan')
plt.xlabel('Berat Badan (kg)')
plt.ylabel('Jumlah')

# Distribusi Tinggi Badan
plt.subplot(1, 3, 3)
sns.histplot(df['Height'], bins=30, kde=True, color='red')
plt.title('Distribusi Tinggi Badan')
plt.xlabel('Tinggi Badan (cm)')
plt.ylabel('Jumlah')

plt.tight_layout()
plt.show()
```

**Distribusi Usia**
* Distribusi miring ke kanan (right-skewed) atau positif skew.

* Mayoritas individu berusia antara 18 hingga 30 tahun, dengan puncaknya di sekitar 21–24 tahun.

* Terdapat ekor panjang ke kanan dengan nilai ekstrem hingga di atas 100 tahun, menunjukkan outlier atau nilai anomali yang perlu ditinjau lebih lanjut.

* Hal ini konsisten dengan boxplot sebelumnya.

**Distribusi Berat Badan**
* Distribusi juga right-skewed.

* Sebagian besar individu memiliki berat antara sekitar 50–150 kg, dengan puncaknya sekitar 90–110 kg.

* Terdapat nilai ekstrem lebih dari 350 kg, yang bisa jadi adalah data salah atau outlier klinis.

* Distribusi ini menunjukkan konsentrasi berat badan di kisaran yang realistis, tetapi ekor panjangnya perlu diperiksa.

**Distribusi Tinggi Badan**
* Distribusi agak mendekati normal, tetapi masih sedikit miring ke kanan.

* Mayoritas tinggi badan berada antara 1.5 hingga 1.9 meter, dengan puncaknya sekitar 1.7 meter.

* Namun, terdapat nilai ekstrem di atas 2.5 meter, yang secara biologis sangat tidak mungkin.

* Ini mengindikasikan kemungkinan kesalahan entri data (input error).
